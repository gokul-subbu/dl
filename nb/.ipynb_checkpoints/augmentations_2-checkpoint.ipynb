```python
image: np.ndarray=cv2.imread(r'e:\images\peta.jpg')
image= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
data: torch.tensor=kornia.image_to_tensor(image)
gauss = kornia.filters.GaussianBlur2d((11, 11), (10.5, 10.5))
x_blur: torch.tensor = gauss(data.float())
img_blur= kornia.tensor_to_image(x_blur.byte())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 10))
axs = axs.ravel()

axs[0].axis('off')
axs[0].set_title('image source')
axs[0].imshow(image)

axs[1].axis('off')
axs[1].set_title('image blurred')
axs[1].imshow(img_blur)
```

```python
def imshow(input: torch.Tensor):
    out: torch.Tensor = torchvision.utils.make_grid(input, nrow=2, padding=5)
    out_np: np.ndarray = kornia.tensor_to_image(out)
    plt.imshow(out_np)
    plt.axis('off')
```

```python
imshow(data)
```

```python
imshow(torch.flip(data, [-2]))
```

```python
def hflip(input: torch.Tensor) -> torch.Tensor:
    return torch.flip(input, [-1])


def vflip(input: torch.Tensor) -> torch.Tensor:
    return torch.flip(input, [-2])


def rot180(input: torch.Tensor) -> torch.Tensor:
    return torch.flip(input, [-2, -1])


def imshow(input: torch.Tensor):
    out: torch.Tensor = torchvision.utils.make_grid(input, nrow=2, padding=5)
    out_np: np.ndarray = kornia.tensor_to_image(out)
    plt.imshow(out_np)
    plt.axis('off')
```

```python
xb_bgr = torch.cat([x_bgr, hflip(x_bgr), vflip(x_bgr), rot180(x_bgr)])
imshow(xb_bgr)
```

```python
xb_rgb = kornia.bgr_to_rgb(xb_bgr)
imshow(xb_rgb)
```

```python
xb_gray = kornia.rgb_to_grayscale(xb_rgb.float() / 255.)
imshow(xb_gray)
```

```python
xb_hsv = kornia.rgb_to_hsv(xb_rgb.float() / 255.)
imshow(xb_hsv[:, 2:3])
```
